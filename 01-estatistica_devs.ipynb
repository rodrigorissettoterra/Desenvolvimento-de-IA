## Conjunto de dados

### Criando o conjunto de dados

Criando um conjunto de dados (dicionário)

```python
dict_medidas = {'idade': [15,18,25,25,40,55,58,60,80], 
                'altura': [160,162,165,168,172,174,174,174,176]}
```

Criando um Dataframe

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
df_medidas
```

### Medidas de posição

Idade

```python
print("A média das idades do dataset, é:",df_medidas['idade'].mean())
print("A mediana das idades do dataset, é:", df_medidas['idade'].median())
print("A moda das idades do datase, é:", df_medidas['idade'].mode())
```

Altura

```python
print("A média das alturas do dataset, é:",df_medidas['altura'].mean())
print("A mediana das alturas do dataset, é:", df_medidas['altura'].median())
print("A moda das alturas do datase, é:", df_medidas['altura'].mode())
```

### Medidas de dispersão

Idade

```python
print("A variância das idades do dataset, é:",df_medidas['idade'].var())
print("O desvio padrão das idades do dataset, é:",df_medidas['idade'].std())
print("O coeficiente de variação das idades do dataset, é:", (df_medidas['idade'].std() / df_medidas['idade'].mean()) * 100, "%.")
```

Altura

```python
print("A variância das alturas do dataset, é:",df_medidas['altura'].var())
print("O desvio padrão das alturas do dataset, é:",df_medidas['altura'].std())
print("O coeficiente de variação das alturas do dataset, é:", (df_medidas['altura'].std() / df_medidas['altura'].mean()) * 100, "%.")
```

### Medidas de forma

Idade

```python
print("A assimetria das idades do dataset, é:",df_medidas['idade'].skew())
print("A curtose das idades do dataset, é:",df_medidas['idade'].kurtosis())
```

Altura

```python
print("A assimetria das alturass do dataset, é:",df_medidas['altura'].skew())
print("A curtose das alturass do dataset, é:",df_medidas['altura'].kurtosis())
```

Describe(), Pandas

Idade

```python
df_medidas['idade'].describe()
```

Altura

```python
df_medidas['altura'].describe()
```

### Correlação

Pearson

```python
df_medidas.corr(method='pearson')
```

Spearman

```python
df_medidas.corr(method='spearman')
```

Observação: Em ambos os coeficientes, podemos notar altíssima correlação entre os dados de Idade e de Altura (acima de 0,9).
